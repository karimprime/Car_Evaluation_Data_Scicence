```python
dataset['Buying_Car']=dataset['Buying_Car'].replace('vhigh',4.0)
dataset['Buying_Car']=dataset['Buying_Car'].replace('high',3.0)
dataset['Buying_Car']=dataset['Buying_Car'].replace('med',2.0)
dataset['Buying_Car']=dataset['Buying_Car'].replace('low',1.0)
dataset.head(10)
```

```python
dataset['Safety_Rate']=dataset['Safety_Rate'].replace('vhigh',4.0)
dataset['Safety_Rate']=dataset['Safety_Rate'].replace('high',3.0)
dataset['Safety_Rate']=dataset['Safety_Rate'].replace('med',2.0)
dataset['Safety_Rate']=dataset['Safety_Rate'].replace('low',1.0)
dataset.tail(1000)
```

```python
dataset['Final_Evaluation']=dataset['Final_Evaluation'].replace('acc',3.0)
dataset['Final_Evaluation']=dataset['Final_Evaluation'].replace('vgood',2.0)
dataset['Final_Evaluation']=dataset['Final_Evaluation'].replace('good',1.0)
dataset['Final_Evaluation']=dataset['Final_Evaluation'].replace('unacc',0.0)
dataset.tail(1000)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset)
dataset=pd.DataFrame(scaler.transform(dataset),columns=['Buying_Car','Number_Doors','Number_of_Persons','Lug_Boot', 'Safety_Rate'])
```

```python
dataset.tail(7)
```

```python
dataset_labels=dataset['Lug_Boot']
dataset=dataset.drop('Lug_Boot',axis=1)
```

```python
from sklearn.model_selection import train_test_split
train_set , test_set=train_test_split(dataset , test_size = 0.2)
train_labels , test_labels=train_test_split(dataset_labels , test_size = 0.2)
print(len(train_set),"train +",len(test_set),"test")
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(train_set,train_labels)
```

```python
from sklearn.metrics import mean_squared_error
dataset_predictions=lin_reg.predict(test_set)
lin_mse=mean_squared_error(test_labels,dataset_predictions)
lin_rmse= np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.metrics import mean_absolute_error
lin_mae=mean_absolute_error(test_labels,dataset_predictions)
lin_mae
```
